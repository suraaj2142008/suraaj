```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\User\Documents\Linear Regression Dataset\Car_Purchasing_Data.csv')
```

```python
df.info()
```

```python
Q1_Annual_Salary = df.Annual_Salary.quantile(0.25)
Q3_Annual_Salary = df.Annual_Salary.quantile(0.75)
IQR_Annual_Salary = Q3_Annual_Salary-Q1_Annual_Salary

lower_limit_Annual_Salary = Q1_Annual_Salary-1.5*IQR_Annual_Salary
upper_limit_Annual_Salary = Q3_Annual_Salary+1.5*IQR_Annual_Salary

a_Annual_Salary = df['Annual_Salary'].mean()

for x in ['Annual_Salary']:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25

    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)

    df.loc[df[x] < min,x] = a_Annual_Salary
    df.loc[df[x] > max,x] = a_Annual_Salary
```
